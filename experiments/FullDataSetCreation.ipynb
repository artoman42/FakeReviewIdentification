# Model Training

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
```

```python
x_train = train_df['tokenized_review']
y_train = train_df['Labels']
```

```python
x_test = test_df['tokenized_review']
y_test = test_df['Labels']
```

```python
class_weights = compute_class_weight(class_weight="balanced", classes=np.unique(y_train), y=y_train)
class_weights_dict = {i: w for i, w in zip(np.unique(y_train), class_weights)}
```

```python
class_weights_dict
```

```python
class_weights
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

```python
pipelineLogReg = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("logreg", LogisticRegression(class_weight="balanced", max_iter=1000)) 
])
```

```python
pipelineLogReg.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred = pipelineLogReg.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
pipelineXGB = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("xgb", XGBClassifier(scale_pos_weight= class_weights, use_label_encoder=False, eval_metric="logloss"))  # Auto-adjusts weights
])
pipelineXGB
```

```python
pipelineXGB.fit(x_train, y_train)
```

```python
y_pred = pipelineXGB.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
y_pred = np.argmax(pipelineXGB.predict_proba(x_test), axis=1)
print(classification_report(y_test, y_pred))
```

```python
pipelineXGB2 = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("xgb", XGBClassifier(scale_pos_weight= class_weights, use_label_encoder=False, eval_metric="logloss")) 
])
```

```python
from sklearn.utils import class_weight
classes_weights = class_weight.compute_sample_weight(
    class_weight='balanced',
    y=train_df['Labels']
)
```

```python
pipelineXGB2.fit(x_train, y_train, xgb__sample_weight=classes_weights)
```

```python
y_pred = pipelineXGB2.predict(x_test)
print(classification_report(y_test, y_pred))
```
