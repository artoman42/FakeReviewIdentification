```python
import instructor
from pydantic import BaseModel
```

```python
import instructor

client = instructor.from_gemini(
    genai.GenerativeModel(
        model_name='models/gemini-1.5-flash'
    )
)
```

```python
def construct_prompt_from_str(input_str:str, placeholder_name:str, value:str) -> str:
    return input_str.replace(placeholder_name, value)
```

```python
prompt = """
There is an enity name - [[placeholder]] from rozetka website, search for this item, and make a 10 completely different reviews about it, and
make them in ukrainan and ukrainian only
Out of 10 reviews: 4 must be positive, 4 must be negative, and 2 neutral
"""
```

```python
class Review(BaseModel):
    review_body:str
```

```python
from typing import List
```

```python
class EntityWithReview(BaseModel):
    entity_name:str
    list_of_reviews:List[Review]
```

```python
def _querry_model(user_prompt:str, response_model:BaseModel) -> EntityWithReview:
    result = client.chat.completions.create(
        response_model=response_model,
        messages = [
            {
                'role':'system',
                'content':'You are proffessional reviewer. You must provide reviews according to the user prompt.'
            },
            {
                'role':'user',
                'content':user_prompt
            }
        ]
    )
    return result
```

```python
results = []
for entity in listofEntities[:5]:
    filled_prompt = construct_prompt_from_str(prompt, '[[placeholder]]', entity)
    entity_result = _querry_model(filled_prompt, EntityWithReview)
    results.append(entity_result)
results    
```
